```python
import string

def wordcount(text):

  try:
    text = text.lower()
    regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
    txt = regex.sub(" " , text)

    words = [w for w in txt.split(" ")
             if (not w in stop) and (len(w) > 3)]
    return len(words)

  except:
    return 0
```

```python
def fill_missing_values(df):
    # df.category_name.fillna(value="missing", inplace=True)
    # df.brand_name.fillna(value="missing", inplace=True)
    df['DESCRIPTION'].fillna(value="missing", inplace=True)
    df['TITLE'].fillna("missing" , inplace = True)
    df['BULLET_POINTS'].fillna("missing" , inplace = True)
    # df['DESCRIPTION'].replace('No description yet',"missing", inplace=True)
    return df
```

```python
train_csv = fill_missing_values(train_csv)
train_csv["is_description_missing"] = np.where(train_csv["DESCRIPTION"] == "missing", 1, 0)
train_csv["is_title_missing"] = np.where(train_csv["TITLE"] == "missing", 1, 0)
train_csv["is_bullet_missing"] = np.where(train_csv["BULLET_POINTS"] == "missing", 1, 0)

train_csv['DESCRIPTION'] = train_csv['DESCRIPTION'].apply(lambda x : lowercase_text(x))
train_csv['TITLE'] = train_csv['TITLE'].apply(lambda x : lowercase_text(x))
train_csv['BULLET_POINTS'] = train_csv['BULLET_POINTS'].apply(lambda x : lowercase_text(x))

#################

test_csv = fill_missing_values(test_csv)
test_csv["is_description_missing"] = np.where(test_csv["DESCRIPTION"] == "missing", 1, 0)
test_csv["is_title_missing"] = np.where(test_csv["TITLE"] == "missing", 1, 0)
test_csv["is_bullet_missing"] = np.where(test_csv["BULLET_POINTS"] == "missing", 1, 0)
test_csv['DESCRIPTION'] = test_csv['DESCRIPTION'].apply(lambda x : lowercase_text(x))
test_csv['TITLE'] = test_csv['TITLE'].apply(lambda x : lowercase_text(x))
test_csv['BULLET_POINTS'] = test_csv['BULLET_POINTS'].apply(lambda x : lowercase_text(x))
```
